### Exercise 4.1: Mean Function


```python
def calculate_mean(values):
    """
    Calculate mean (average) of a list.

    Args:
        values: List of numbers

    Returns:
        Mean as float
    """
    # YOUR CODE HERE
    mean = sum(values)/len(values) *(1.00)
    return mean


# Test
test = [10, 20, 30, 40, 50]
result = calculate_mean(test)
print(f"Mean of {test}: {result}")
```

```python
assert abs(calculate_mean([1, 2, 3, 4, 5]) - 3.0) < 0.01
assert abs(calculate_mean([10, 20, 30]) - 20.0) < 0.01
print("✓ Exercise 4.1 passed!")
```

### Exercise 4.2: Distance Function


```python
def euclidean_distance(point1, point2):
    """
    Calculate Euclidean distance between two 2D points.

    Formula: sqrt((x1-x2)² + (y1-y2)²)
    """
    # YOUR CODE HERE
    x1, y1 = point1
    x2, y2 = point2

    dist = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return dist

# Test
p1, p2 = [0, 0], [3, 4]
dist = euclidean_distance(p1, p2)
print(f"Distance: {dist}")
```

```python
assert abs(euclidean_distance([0, 0], [3, 4]) - 5.0) < 0.01
assert abs(euclidean_distance([0, 0], [1, 1]) - 1.414) < 0.01
print("✓ Exercise 4.2 passed!")
```

### Exercise 4.3: Normalization Function


```python
def normalize_list(values):
    """
    Normalize to [0,1] using min-max scaling.

    Formula: (x - min) / (max - min)
    """

    # YOUR CODE HERE
    Min = min(values)
    Max = max(values)
    range = Max - Min

    normalized = []
    for val in values:
      normal_val = (val - Min) / range
      normalized.append(normal_val)
    return normalized

# Test
data = [10, 20, 30, 40, 50]
result = normalize_list(data)
print(f"Normalized: {[f'{x:.2f}' for x in result]}")
```

```python
r = normalize_list([0, 50, 100])
assert abs(r[0]-0.0) < 0.01 and abs(r[1]-0.5) < 0.01 and abs(r[2]-1.0) < 0.01
print("✓ Exercise 4.3 passed!")
```

### Exercise 4.4: Sigmoid Function


```python
import numpy as np
def simple_sigmoid(x):
    """
    Sigmoid: σ(x) = 1 / (1 + e^(-x))
    """
    # YOUR CODE HERE
    return 1 / (1 + np.exp(-x))


# Test
for x in [-5, 0, 5]:
    print(f"sigmoid({x}) = {simple_sigmoid(x):.4f}")
```

```python
assert abs(simple_sigmoid(0) - 0.5) < 0.01
assert simple_sigmoid(10) > 0.9 and simple_sigmoid(-10) < 0.1
print("✓ Exercise 4.4 passed!")
```

```python
# Given data
data_points = [1.5, 4.2, 2.8, 5.6, 3.1]
true_labels = [0, 1, 0, 1, 1]

# Task 1: Make predictions (if > 3.0, predict 1, else 0)
def make_predictions(data_points):
  prediction =[]
  for i in range(len(data_points)):
    prediction.append(int(data_points[i] > 3.0))
  return prediction


# Task 2: Calculate accuracy
# YOUR CODE HERE
def accuracy(predictions, true_labels):
  correct = 0
  for i in range(len(predictions)):
    if predictions[i] == true_labels[i]:
      correct += 1
  accuracy = (correct / len(predictions)) * 100
  return accuracy

# Task 3: Normalize data
# YOUR CODE HERE
def normalized(data_points):
  normalized = []
  min_val = min(data_points)
  max_val = max(data_points)
  range_val = max_val - min_val
  for val in data_points:
    norm_val = (val - min_val) / range_val
    normalized.append(norm_val)
  return normalized

normalized_data = normalized(data_points)
predictions = make_predictions(data_points)
accuracy = accuracy(predictions, true_labels)

# Results
print("╔═══════════════════════════════╗")
print("║   ML Pipeline Results        ║")
print("╚═══════════════════════════════╝")
print(f"\nData: {data_points}")
print(f"Normalized: {[f'{x:.3f}' for x in normalized_data]}")
print(f"\nTrue: {true_labels}")
print(f"Pred: {predictions}")
print(f"\nAccuracy: {accuracy:.1f}%")
```

```python
assert predictions == [0, 1, 0, 1, 1] and accuracy == 100.0
assert len(normalized_data) == 5 and all(0 <= x <= 1 for x in normalized_data)
print("✓ Final challenge passed!")
print("\n🎉 You built a complete ML pipeline!")
```
